```python
#dataframe 3
#read
decennial_df = pd.read_csv("decennial_pr_202.csv")

# eliminate NaN rows & PR row
decennial_df = decennial_df.dropna().drop([157])

#renaming columns / cleaning names
decennial_df.columns = [c.lower().replace('total:!!', '').replace(' ', '_').replace('label_(grouping)','municipio').replace(':', '').replace(':!!', '').replace('!!', '') for c in decennial_df.columns]

# Replacing str values for another str
decennial_df['municipio'] = decennial_df['municipio'].str.replace('Municipio, Puerto Rico', '').str.strip()

decennial_df


```

```python
merged_df = pd.merge(merged,decennial_df[['municipio', 'total']],on='municipio')
merged_df
```

```python
merged_df.rename(columns={'2020_pop': 'pop2020', '2017_pop': 'pop2017', '2018_pop':'pop2018', '2019_pop':'pop2019'}, inplace=True)
merged_df
```

```python
merged_df.dtypes
```

```python
# Turning dtype[0] into int 

merged_df['pop2017'] = merged_df['pop2017'].str.replace(',', '')
merged_df['pop2018'] = merged_df['pop2018'].str.replace(',', '').astype(int)
merged_df['pop2019'] = merged_df['pop2019'].str.replace(',', '').astype(int)
merged_df['pop2020'] = merged_df['pop2020'].str.replace(',', '').astype(int)
merged_df.dtypes
```

```python
#Creating a new column for housing damages
merged_df['per_capita_damages'] = merged_df.estimated_housing_damages_ / merged_df.pop2017 
merged_df
```

```python
#Creating a new column for pending needs per capita
merged_df['per_capita_needs'] = merged_df.pending_housing_needs / merged_df.pop2020 
merged_df
```
